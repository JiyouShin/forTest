```python
import math
import numpy as np
```

```python
def solution(id_list, report, k):
    name_list = []
    for i in list(dict.fromkeys(report)): 
        split = i.split()
        for i in split: 
            name_list.append(i)
    
    reporter = []
    reported = []

    count = 0
    for i in name_list: 
        if count % 2 == 0 :
            reporter.append(i)
        else : 
            reported.append(i)
        count += 1
    
    # print("reporter :: ", reporter)
    # print("reported :: ", reported)

    reported_count_dict = {}
    for i in reported:
        try : reported_count_dict[i] += 1
        except : reported_count_dict[i] = 1
    
    # print("reported_count_dict :: ", reported_count_dict)

    np_reporter = np.array(reporter)
    mail_count = {}

    for a in range(0, len(id_list)):
        reporter_index = np.where(np_reporter==id_list[a])

        if len(reporter_index[0]) == 0:
            mail_count[id_list[a]] = 0

        for b in reporter_index[0]:
            if reported_count_dict[reported[b]] >= k:
                try : mail_count[id_list[a]] += 1
                except : mail_count[id_list[a]] = 1
            else : 
                try : mail_count[id_list[a]] += 0
                except : mail_count[id_list[a]] = 0


    answer = list(mail_count.values())

    return answer
```

```python
solution(	["muzi", "frodo", "apeach", "neo"], ["muzi frodo", "apeach frodo", "frodo neo", "muzi neo", "apeach muzi"], 2)
```
