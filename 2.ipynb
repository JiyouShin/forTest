```python
def solution(lottos, win_nums) : 
    dict = {7-i: i for i in range(1, 7)}
    dict.update({0 : 6})

    # print(dict)

    same_count = 0
    for a in lottos:
        for b in win_nums:
            if a == b :
                same_count += 1
    
    # print("same_count :: ", same_count)

    max_count = 0

    for i in lottos :
        if i == 0 :
            max_count += 1
    
    answer = [dict[same_count + max_count], dict[same_count]]

    return answer
```

```python
solution([0, 0, 0, 0, 34, 20], [11, 12, 14, 42, 34, 20])
```
